```python
engine=train['기통']
engine
```

```python
def tf_df(train, mode='train'):

    model_name = train['모델명']
    years = train['연식']

    tf_model_name = []
    tf_years = []

    for name, year in zip(model_name, years):
        tf_name = name.split(' ')[0]
        tf_model_name.append(tf_name)

        tf_year = 2021 - year
        tf_years.append(tf_year)

    distance = train['주행거리']
    distance.head()

    tf_distance=[]
    for d in distance:
        if 'ml' in d:
            if '만' in d:
                val = int(d.split('만')[0])
                val = val*10000*1.609344

        elif 'km' in d:
            if '만' in d:
                val = int(d.split('만')[0])
                val = val * 10000
            elif '천' in d:
                val = int(d.split('천')[0])
                val = val * 1000
            else: 
                val = int(d.split('km')[0])

        tf_distance.append(val)
    len(tf_distance)

    power=train['최대출력(마력)']
    torq=train['최대토크(kgm)']
    fuel=train['연료']
    method=train['구동방식']
    engine=train['기통']
    
    if mode=='train':
        sale=train['가격(만원)']

        df = pd.DataFrame({
                             #'model_name': tf_model_name,
                             'sale': sale,

                             'years': tf_years,
                             'distance': tf_distance,
                             'power':power,
                             'torq': torq,
                             'fuel': fuel,
                             'method': method,
                             'engine': engine

        })
        
    else:
         df = pd.DataFrame({
                             #'model_name': tf_model_name,
                             'years': tf_years,
                             'distance': tf_distance,
                             'power':power,
                             'torq': torq,
                             'fuel': fuel,
                             'method': method,
                             'engine': engine
        })
            
    return df
```

```python
df = tf_df(train)
test_df = tf_df(test, mode='test')
```

```python
df.head()
```

```python
test_df.head()
```

```python
training_df = pd.get_dummies(df)
a = pd.get_dummies(df['engine'])
print(a.shape)

for c in a.columns:
    training_df.insert(5, c, a[c])
training_df=training_df.drop(['engine'], axis=1)
training_df = training_df.fillna(training_df.mean())
#training_df.mean()
training_df
```

```python

numeric = ['sale', 'years', 'distance', 'power', 'torq']

for col in numeric:
    tf_df = training_df[col]
    tf_df = (tf_df - tf_df.min()) / (tf_df.max() - tf_df.min())
    training_df[col]=tf_df
training_df
```

```python
y = training_df['sale']
x = training_df.drop(['sale'], axis=1)
x
```

# model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)

accuracy = model.score(x, y)
print(accuracy)
```

```python
y_hat= model.predict(x)
```
